```python
import pandas as pd
import matplotlib.pyplot as plt

#reading the database
data = pd.read_csv("stock_data.csv")

#scatter plot with day against tip
plt.scatter(data['Open'], data['Close'])

#Adding Title to the plot
plt.title("Scatter Plot")

#Setting the X and Y labels
plt.xlabel('Open')
plt.ylabel('Close')

plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

#reading the database
data = pd.read_csv("stock_data.csv")

#scatter plot with day against tip
plt.scatter(data['Open'], data['Low'], c=data['Low'],
            s=data['Close'])

#Adding Title  to the plot
plt.title("Scatter Plot")

#setting the x and y labels
plt.xlabel('Open')
plt.ylabel('Low')

plt.colorbar()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# reading the database
data = pd.read_csv("stock_data.csv")

# Scatter plot with day against tip
plt.plot(data['Open'])
plt.plot(data['Low'])

# Adding Title to the Plot
plt.title("Scatter Plot")

# Setting the X and Y labels
plt.xlabel('Day')
plt.ylabel('Tip')
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

#reading the database
data = pd.read_csv("stock_data.csv")

#Bar chart with day against tip
plt.bar(data['Date'], data['Open'])

plt.title("Bar Chart")

#Setting the X and Y labels
plt.xlabel('Date')
plt.ylabel('Open')

#Adding the legends
plt.show()
```
