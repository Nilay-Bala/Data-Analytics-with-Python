```python
#importing pandas as pd
import pandas as pd

#Creating the DataFrame
df = pd.DataFrame({"A": [1, 2, 3, None, 5],
                   "B": [6, 7, 8,9, None],
                   "C": [10, 11, 12, 13, 14],
                   "D": [15, 16, None, 18, 19]})

#Creating the index
index = ['Row_1', 'Row_2', 'Row_3', 'Row-4', 'Row_5']

#set the index
df.index = index

#print the DataFrame
df
```

```python
#use dataframe.transform() function to add 10 to each element of the dataframe.
result = df.transform(func = lambda x : x + 10)
result
```

```python
#use DataFrame.transform() function to find the square root and the result of euler's number raised to each element of the dataframe
import pandas as pd

#creating the dataframe
df = pd.DataFrame({"A": [1, 2, 3, None, 5],
                   "B": [6, 7, 8,9, None],
                   "C": [10, 11, 12, 13, 14],
                   "D": [15, 16, None, 18, 19]})

#create the index
index = ['Row_1', 'Row_2', 'Row_3', 'Row_4', 'Row_5']

#Set the index
df.index = index

#print the DataFrame
print(df)
```

```python
result = df.transform(func = ['sqrt', 'exp'])
result
```

```python
text = 'HiI mY nAmE iS NIlaY'

print("The string is : ")
print(text)

#upper() function to convert string to upper case
print("\nConverted String to upper case: ")
print(text.upper())

#lower() function to convert string to lower case
print("\nConverted String to lower case: ")
print(text.lower())

#title() function Converts the first character to upper case and rest to lower case
print("\nConverted first letter to upper in every word: ")
print(text.title())

#swaps the case of all characters in the string upper case character to lower case and viceversa
print("\nString Case Swapping: ")
print(text.swapcase())

#convert the frist character of a string to uppercase
print("\nConverted String first letter to upper: ")
print(text.capitalize())

#original string never changes
print("\nOriginal String: ")
print(text)
```
