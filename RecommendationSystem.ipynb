```python
import pandas as pd

# Replace 'IMDb Movies India.csv' with the actual filename of your dataset
dataset_file = 'IMDb Movies India.csv'

# Load the dataset using pandas with the appropriate encoding
movies_df = pd.read_csv(dataset_file, encoding='latin-1')

# Display the first few rows of the dataset to ensure it was loaded correctly
movies_df.head()
```

```python
# Calculate count rating of all movies
movies_df.groupby('Name')['Rating'].count().sort_values(ascending=False).head()
```

```python
# creating dataframe with 'rating' count values
ratings = pd.DataFrame(movies_df.groupby('Name')['Rating'].mean())

ratings['num of ratings'] = pd.DataFrame(movies_df.groupby('Name')['Rating'].count())

ratings.head()
```

```python
# Create the pivot table representing user ratings for movies
moviemat = movies_df.pivot_table(index=None, columns='Name', values='Rating')

# Display the pivot table
print(moviemat)
```

```python
# Sorting values according to
# the 'num of rating column'
moviemat = movies_df.pivot_table(index =None,
			columns ='Name', values ='Rating')

moviemat.head()

ratings.sort_values('num of ratings', ascending = False).head(10)
```

```python
# analysing correlation with similar movies
Anjaam_user_ratings = moviemat['Anjaam']
Baazi_user_ratings = moviemat['Baazi']

Anjaam_user_ratings.head()
```

```python
# Step 1: Calculate the correlation between movies based on their ratings
movie_correlation = moviemat.corr()

# Step 2: Display the correlation matrix
print(movie_correlation)
```

```python
movie1 = 'Guru'
movie2 = 'Dushman'

# Step 2: Extract the user ratings for both movies from the 'moviemat' pivot table
movie1_user_ratings = moviemat[movie1]
movie2_user_ratings = moviemat[movie2]

# Step 3: Calculate the correlation between the two movies' ratings
correlation = movie1_user_ratings.corr(movie2_user_ratings)

# Display the correlation value
print(f"Correlation between '{movie1}' and '{movie2}': {correlation}")
```

```python
#Similar movies like Guru
# Step 1: Sort the movies in 'moviemat' based on correlation with "Guru" in descending order and display the top 10 similar movies
print(moviemat.corrwith(moviemat['Guru']).sort_values(ascending=False).head(10))

# Step 2: Create the 'corr_Guru' DataFrame with the correlation values for "Guru" and other movies
corr_Guru = pd.DataFrame(moviemat.corrwith(moviemat['Guru']), columns=['Correlation'])
corr_Guru.dropna(inplace=True)

# Step 3: Join the 'num of ratings' column from the 'ratings' DataFrame to the 'corr_Guru' DataFrame
corr_Guru = corr_Guru.join(ratings['Rating'])

# Step 4: Display the head of the 'corr_Guru' DataFrame, which now includes the 'num of ratings' column
print(corr_Guru.head())

# Step 5: Filter the DataFrame to consider only movies with more than 3 ratings and display the top few similar movies
filtered_corr_Guru = corr_Guru[corr_Guru['Rating'] >= 3]
print(filtered_corr_Guru.sort_values('Correlation', ascending=False).head())
```
