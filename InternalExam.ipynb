```python
# (1) Sort the list using bubble sort and selection sort

import pandas as pd

def bubble_sort(arr):
    n = len(arr)
    for i in range(n):  # Loop through the list for each pass
        for j in range(0, n-i-1):  # Loop through unsorted portion of the list
            if arr[j] > arr[j+1]:  # If current element is greater than the next element
                arr[j], arr[j+1] = arr[j+1], arr[j]  # Swap the elements

# Example list
my_list = [34, 23, 45, 67, 2, 10, 80, 21, 12]
print(my_list)
bubble_sort(my_list)
print("Sorted list using Bubble Sort:", my_list)

def selection_sort(arr):
    n = len(arr)
    for i in range(n):  # Loop through the list
        min_index = i  # Assume the minimum element is at index i
        for j in range(i+1, n):  # Loop through unsorted portion of the list
            if arr[j] < arr[min_index]:  # If current element is smaller than the assumed minimum
                min_index = j  # Update the index of the minimum element
        arr[i], arr[min_index] = arr[min_index], arr[i]  # Swap the minimum element with the current element

my_list = [34, 23, 45, 67, 2, 10, 80, 21, 12]
selection_sort(my_list)
print("Sorted list using Selection Sort:", my_list)

```

```python
# (2) Write the pyhton code for fibonacci series 

#Itterative approach
def fibonacci_iterative(n):
    fib_series = [0, 1]
    for i in range(2, n):
        next_fib = fib_series[i - 1] + fib_series[i - 2]
        fib_series.append(next_fib)
    return fib_series

# Generate the first 10 Fibonacci numbers using the iterative approach
n = 10
fib_series = fibonacci_iterative(n)
print("Fibonacci Series (Iterative):", fib_series)

#Recursive approach
def fibonacci_recursive(n):
    if n <= 0:
        return []
    elif n == 1:
        return [0]
    elif n == 2:
        return [0, 1]
    else:
        fib_series = fibonacci_recursive(n - 1)
        next_fib = fib_series[-1] + fib_series[-2]
        fib_series.append(next_fib)
        return fib_series

# Generate the first 10 Fibonacci numbers using the recursive approach
n = 10
fib_series = fibonacci_recursive(n)
print("Fibonacci Series (Recursive):", fib_series)


```

```python
# (6) Sort the specified tuple in descending order:
#("h", "b", "a", "c", "f", "d", "e", "g")

#descending order
my_tuple = ("h", "b", "a", "c", "f", "d", "e", "g")
sorted_tuple_descending = tuple(sorted(my_tuple, reverse=True))

print("Sorted Tuple in Descending Order:", sorted_tuple_descending)

#Ascending order
sorted_tuple_ascending = tuple(sorted(my_tuple))
print("Sorted Tuple in Ascending Order:", sorted_tuple_ascending)
```

```python
#(i)Replacing the NaN with the column mean
#compute mean
c = avg = 0
for ele in df['Marks']:
    if str(ele).isnumeric():
        c += 1
        avg += ele
        avg /= c
        
#Replace missing values
df = df.replace(to_replace="NaN", value=avg)

#Display data
df
```

```python
#(ii) replace the Gender column data by categorizing them into different numbers.

#Data Replacing in Data Wrangling
#in the Gender column, we can replace the Gender column data by categorizing them into different numbers.
#Categorizing gender
df['Gender'] =df['Gender'].map({'M': 0, 'F': 1,}).astype(float)

#Display data
df
```

```python
#plot a Bar chart and scatter chart on Gender vs marks.

import pandas as pd
import matplotlib.pyplot as plt

# Creating a DataFrame
df = pd.DataFrame({'Name': ['Jai', 'Princi', 'Gaurav', 'Anuj', 'Ravi', 'Natasha', 'Riya'],
                   'Age': [17, 17, 18, 17, 18, 17, 17],
                   'Gender': ['M', 'F', 'M', 'M', 'M', 'F', 'F'],
                   'Marks': [90, 76, None, 74, 65, None, 71]})

# Plotting a Bar Chart
gender_marks_mean = df.groupby('Gender')['Marks'].mean()
gender_marks_mean.plot(kind='bar', color=['blue', 'pink'])
plt.title('Gender vs Average Marks')
plt.xlabel('Gender')
plt.ylabel('Average Marks')
plt.show()

# Plotting a Scatter Plot
plt.scatter(df['Gender'], df['Marks'], color='green', marker='o')
plt.title('Gender vs Marks')
plt.xlabel('Gender')
plt.ylabel('Marks')
plt.show()
```
